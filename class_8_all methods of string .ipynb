```python
# Sample string
s = "Hello, WorLd!"

# String methods
print(s.capitalize())       # Capitalize the first character
print(s.casefold())         # Convert to lowercase
print(s.center(21, '*'))    # Center the string with a specified width and fill character
print(s.count('o'))         # Count occurrences of a substring
print(s.encode())           # Encode the string
print(s.endswith('!'))      # Check if the string ends with a specified suffix
print(s.expandtabs(4))      # Expand tabs in the string
print(s.find('World'))      # Find the position of a substring
print(s.format())           # Format the string
print(s.index('WorLd'))     # Find the position of a substring (raises ValueError if not found)
print(s.isalnum())          # Check if all characters are alphanumeric
print(s.isalpha())          # Check if all characters are alphabetic
print(s.isascii())          # Check if all characters are ASCII
print(s.isdecimal())        # Check if all characters are decimals
print(s.isdigit())          # Check if all characters are digits
print(s.isidentifier())     # Check if the string is a valid identifier
print(s.islower())          # Check if all characters are lowercase
print(s.isnumeric())        # Check if all characters are numeric
print(s.isprintable())      # Check if all characters are printable
print(s.isspace())          # Check if all characters are whitespace
print(s.istitle())          # Check if the string is title-cased
print(s.isupper())          # Check if all characters are uppercase
print('-'.join(['Hello', 'World']))  # Join iterable elements with a separator
print(s.ljust(20, '*'))     # Left justify the string with a specified width and fill character
print(s.lower())            # Convert to lowercase
print(s.lstrip('H'))        # Remove leading characters
print(s.partition(','))     # Partition the string into three parts
print(s.removeprefix('Hel')) # Remove prefix
print(s.removesuffix('!'))  # Remove suffix
print(s.replace('World', 'Python'))  # Replace occurrences of a substring
print(s.rfind('o'))         # Find the last occurrence of a substring
print(s.rindex('o'))        # Find the last occurrence of a substring (raises ValueError if not found)
print(s.rjust(20, '*'))     # Right justify the string with a specified width and fill character
print(s.rpartition(','))    # Partition the string into three parts from the right
print(s.rsplit(','))        # Split the string from the right
print(s.rstrip('!'))        # Remove trailing characters
print(s.split(','))         # Split the string
print(s.splitlines())       # Split the string at line breaks
print(s.startswith('Hello'))# Check if the string starts with a specified prefix
print(s.strip('H!'))        # Remove leading and trailing characters
print(s.swapcase())         # Swap case of all characters
print(s.title())            # Convert to title case
print(s.upper())            # Convert to uppercase
```
